```python
# a) Strip whitespace from 'Country/Territory'
gdp_df['Country/Territory'] = gdp_df['Country/Territory'].str.strip()
```

```python
# Strip leading/trailing whitespace from all string columns
gdp_df = gdp_df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
```

```python
# b) Convert 'IMF_Year' to datetime
gdp_df['IMF_Year'] = pd.to_datetime(
   gdp_df['IMF_Year'],
   format='%Y',
   errors='coerce'
)
```

```python
#Check the datatype first for another column.
# should be an integer. #Check Dataframe it does show numbers.
print(df['WorldBank_Estimate'].dtype)
```

```python
# c) Convert 'WorldBank_Estimate' to float with 1 decimal
gdp_df['WorldBank_Estimate'] = (
   pd.to_numeric(gdp_df['WorldBank_Estimate'], errors='coerce')
     .fillna(0)
     .astype(float)
     .round(1)
)
```

```python
#Add on the end for 2 decimal places:
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# c) Convert 'WorldBank_Estimate' to interger
gdp_df['WorldBank_Estimate'] = pd.to_numeric(gdp_df['WorldBank_Estimate'], errors='coerce').fillna(0).astype(int)
```

```python
col = 'IMF_Estimate'
Q1 = gdp_df[col].quantile(0.25)
Q3 = gdp_df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = gdp_df[(gdp_df[col] < lower) | (gdp_df[col] > upper)]
```
